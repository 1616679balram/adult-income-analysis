```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cleaned_adult.csv")

print("Dataset shape before outlier removal:", df.shape)

df.head()
```

```python
numeric_cols = [
    "age",
    "education_num",
    "hours_per_week",
    "capital_gain",
    "capital_loss"
]

df[numeric_cols].describe()
```

```python
def remove_outlier_iqr(data, column):

    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"\nColumn: {column}")
    print("Lower bound:", lower_bound)
    print("Upper bound:", upper_bound)

    filtered = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

    print("Rows remaining:", filtered.shape[0])

    return filtered
```

```python
df_no_outliers = df.copy()

for col in numeric_cols:
    df_no_outliers = remove_outlier_iqr(df_no_outliers, col)

print("\nFinal dataset shape after outlier removal:", df_no_outliers.shape)
```

```python
print("Rows before cleaning:", df.shape[0])
print("Rows after outlier removal:", df_no_outliers.shape[0])
```

```python
df_no_outliers.to_csv("final_adult_cleaned.csv", index=False)

print("Final cleaned dataset saved as final_adult_cleaned.csv")
```
